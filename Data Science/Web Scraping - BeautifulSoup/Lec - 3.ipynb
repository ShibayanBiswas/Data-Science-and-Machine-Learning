##Books to Scrape##

```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://books.toscrape.com/')
response
```

```python
response.headers
```

```python
html_data = response.text
html_data
```

```python
data = BeautifulSoup(html_data, 'html.parser')
data
```

```python
print(data.prettify())
```

```python
data.title
```

```python
data.title.string
```

```python
data.a
```

```python
data.a.string
```

```python
data.header.a
```

```python
data.header.a.string
```

```python
book_1 = data.find(class_ = 'product_pod')
book_1
```

```python
book_1.h3.a
```

```python
book_1.h3.a['title']
```

```python
book_1.h3.a['href'] # This is the relative url
```

```python
print('http://books.toscrape.com/' + book_1.h3.a['href'])
```

```python
books = data.find_all(class_ = 'product_pod')
base_url = 'http://books.toscrape.com/'
books_url = []
for i in books:
  books_url.append(base_url + i.h3.a['href'])
books_url
```
