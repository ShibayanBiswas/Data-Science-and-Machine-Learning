## Lecture - 3 (Webelement Methods and Properties)

Let's try to scrape data from the website : 'http://kanview.ks.gov/PayRates/PayRates_Agency.aspx'.

This website provides us the data of the "Employee Compensation" for different agencies.

There is a list of different agencies available on the website. The data we get is for the year 2018. 

Let's click on a specific agency say "Adjutant General". Our browser gets redirected to the webpage  of the agency "Adjutant General". The url is the same as that of the "Home" page.

We can now see the list of "Job Title" for the agency "Adjutant General". Each "Job Title" contains the information of all the employees under it.

This website does not contain "normal links". It contains "JavaScript links". We cannot directly make a get() request to "JavaScript links". Here "BeautifulSoup" fails eventually. Hence, "Selenium" is the only option.

Suppose we want to find the details of all the employees that come under the agency "Adjutant General".

#### Approach - 1 (normal)

```python
# importing the necessary libraries
from selenium import webdriver

# initiating a web driver session (for Google Crome browser)
driver = webdriver.Chrome(executable_path = "/path/to/chromedriver")
```

```python
# visting the website : 'http://kanview.ks.gov/PayRates/PayRates_Agency.aspx'
driver.get('http://kanview.ks.gov/PayRates/PayRates_Agency.aspx')
```


On inspecting the webpage, we see that there are 72 "Job Title".

Read the documentation of "pandas.read_html". It's returntype is a list.

```python
# finding the agency link and clicking it
agency = driver.find_element(By.LINK_TEXT, 'Adjutant General')
agency.click()

# extracting the employee details (table) for each "Job Title" in the agency "Adjutant General"
from bs4 import BeautifulSoup
import pandas as pd

dataFrameList = []    # here we will append the employee details (table) for each tob title

for i in range(73):
    # extracting the link of each job title and clicking it (inspect the webpage of the agency "Adjutant General")
    link = driver.find_element(By.ID, 'MainContent_uxLevel2_JobTitles_uxJobTitleBtn_' + str(i))
    link.click()
    
    # parsing the content of the webpage for a given job title link
    data = BeautifulSoup(driver.page_source, 'lxml')
    
    # extarcting the employee details (table) and storing it in the variable "dataFrameList"
    emp_table = data.table
    df = pd.read_html(str(emp_table), header = 0)    # "header = 0" means make the 0th row as the column header for each dataframe
    dataFrameList.append(df[0])
    
    # to click on the back button
    driver.back()
```

```python
# close the webdriver session
driver.close()
```

#### Approach - 2 (using Exception)

```python
# importing the necessary libraries
from selenium import webdriver

# initiating a web driver session (for Google Crome browser)
driver = webdriver.Chrome(executable_path = "/path/to/chromedriver")
```

```python
# visting the website : 'http://kanview.ks.gov/PayRates/PayRates_Agency.aspx'
driver.get('http://kanview.ks.gov/PayRates/PayRates_Agency.aspx')
```

```python
# finding the agency link and clicking it
agency = driver.find_element(By.LINK_TEXT, 'Adjutant General')
agency.click()

# importing the exception, and then using try-except structure
from selenium.common.exceptions import NoSuchElementException

dataFrameList = []    # here we will append the employee details (table) for each tob title

i = 0    # iterator variable

while True:
    # execute these set of statements until we get the "NoSuchElementException" Exception
    try:
        # extracting the link of each job title and clicking it (inspect the webpage of the agency "Adjutant General")
        link = driver.find_element(By.ID, 'MainContent_uxLevel2_JobTitles_uxJobTitleBtn_' + str(i))
        link.click()

        # parsing the content of the webpage for a given job title link
        data = BeautifulSoup(driver.page_source, 'lxml')
        
        # extarcting the employee details (table) and storing it in the variable "dataFrameList"
        emp_table = data.table
        df = pd.read_html(str(emp_table), header = 0)
        dataFrameList.append(df[0])
        
        # to click on the back button
        driver.back()
        
        # incrementing the iterator variable for generating the next link
        i += 1
    
    # execute these set of statements when we get the "NoSuchElementException" Exception (for the last "Job Title")
    except NoSuchElementException:
        break
```

Here from Approach - 1 and Approach - 2, we get a list of  dataframes for each of the "Job Title" under the agency "Adjutant General". 

Now if we want to merge all the dataframes into a single dataframe containg the details of all the "Employee" from all the "Job Title".

We can take the help of the method "pandas.concat()"  to merge multiple dataframes.

```python
pd.concat(dataFrameList, ignore_index = True)    # "ignore_index = True" means make the row indexes aligned i.e. continious from 0 - (n - 1)
```

```python
# close the webdriver session
driver.close()
```

## Lecture - 4 (locating elements - 2)

In this session, we will discuss about the attributes : 

• find_element(By.XPATH, '') - It is used to loacte and extract data from a webpage using the XML Path.

• find_element(By.CSS_SELECTOR, '') - It is used to loacte and extract data from a webpage using the CSS Selector.

These attributes are used to locate and extract data from a webpage.

### Selecting Nodes :

• nodename - Selects all nodes with the name "nodename".

• / - Selects from the root node.

• // - Selects nodes in the document from the 
current node that match the selection no matter where they are.

• @ - Selects attributes.

We can use this with the help of the syntax : 
#### //tag[@attribute = ‘value’] 



HTML is a kind of XML document. So, we can Tke the help of "XPATH" in the HTML document to reach our final destination.

We try to create a path with the help of which we try to reach the required node. 

We create a path with the help of different nodes and attributes.

"XPATH" can be of 2 different types - absolute and relative.

### Partial	Match :

• //tag[contains(@attribute, ‘value’)] - Helpful when there are multiple classes.

• //tag[starts-with(@attribute, ‘value’)

Let's scrape the website : ''http://books.toscrape.com''

```python
# importing the necessary libraries
from selenium import webdriver

# initiating a web driver session (for Google Crome browser)
driver = webdriver.Chrome(executable_path = "/path/to/chromedriver")
```

```python
# visting the website : 'https://books.toscrape.com/'
driver.get('https://books.toscrape.com/')
```

```python
# extracting the title of the webpage using "XPATH"
from selenium.webdriver.common.by import By

title = driver.find_element(By.XPATH, '/html/head/title')
```

In order to verify waht we have got, we can use the attributes : "outerHTML", "innerHTML".

```python
# verifying outerHTML
title.get_attribute('outerHTML')
```

```python
# verifying innerHTML
title.get_attribute('innerHTML')
```

Hence, we can verify that we have extracted the "title" of the webpage using "XPATH" successfully.

Note : To locate and extract an element from a HTML webpage using, "XPATH", first visit the webpage. Inspect the HTML coding through Google Chrome. Now from the inspect menu, click on the "select element" button and with the help of the mose pointer select the desired element from the webpage and click on it. Now, you will get the tag corresponding to the desired element on the HTML webpage. Right click on the tag --> Copy --> full XPath.

Now we can used the "XPATH" we copied in our code to extract the desired element from the HTML webpage.

```python
# verifying outerHTML
title.get_attribute('outerHTML')
```

## Lecture - 5 (Type into fields)

in this lecture we will see how to type some text inside a textbox present in some website.

For example : Let's search the word "Selenium" on Google.

```python
# importing the necessary libraries
from selenium import webdriver

# initiating a web driver session (for Google Crome browser)
driver = webdriver.Chrome(executable_path = "/path/to/chromedriver")
```

```python
# visting the website : 'https://www.google.com/'
driver.get('https://www.google.com/')
```

In order to pass some text in a searchbox present in a webpage, we first have to locate the searchbox and with the help of the method "send_keys()", we pass the text in the searchbox and with the help of the function "submit()" , we can submit the text in the searchbox.

```python
# locating the textbox and searching for the word "Selenium"
text_box = driver.find_element(By.NAME, 'q')
text_box.send_keys('Selenium')
text_box.submit()
```

Here we can see that the webdriver has redirected us to the search results for the keyword "Selenium".

```python
# to click on the back button
driver.back()
```
