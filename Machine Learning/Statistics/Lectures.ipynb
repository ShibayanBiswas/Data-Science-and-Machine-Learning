## Lecture - 2 (Population and Sample)

Now we will see how can we select the data over which we can apply our analysis to get answers of our different questions.

For example: A company which wants to buy 10 thousand bulbs from an outside manufacturer and and the condition is that it will buy the set of bulbs only if there are less than 1% defective bulbs present in that particular set of bulbs.

To do this, the company can take the 10 thousand bulbs from the manufaturer and it can perform a check over each and every bulb to know what proportion of bulbs are defective. Then they can bome up with the conclusion whether they want to buy the set of bulbs or not. 

In this method there is a lot of effort to be applied from the company side as they have to perform a check over 10 thousand bulbs which is not at all effective in terms of cost and time. Hence it has to find a better solution in which their time and effort is minimised. 

So, rather than performing a check on all the 10 thousand bulbs, the company can randomly select 500 bulbs and it can perform a check on these 500 bulbs. Based on the result obtained, they can figure out what proportion of bulbs are defective among the 10 thousand bulbs. 

We actually follow this rule in many real world scenarios. Whenever we anser a question related to a large set of data, we generally take a subset of that data and we apply our analysis over that subset and then we generalise our result for the large st of data.

Here the 2 proportions of data actually have a term in statistics. The large set of data is termed as "Populaton" and the subset of the large set of data over which we apply our analysis is called "Sample".

Population & Sample : 

• A set of all individuals relevant to a particular statistical question is called a population.

• A smaller group selected from a population is called a sample.

Suppose I want to figure out the average marks obtained by students of class "10" in the year "2019" from "Delhi". In this case the "Population" should be all the students in "Delhi" who are currently studying in calss "10". The "Sample" should be a subset of the "Population" and then we can find out the average. Then we can generalize the average calculated for the "Population".

There are 2 more terms which are important in "Statistics". They are :

1) Parameter - Descriptive measure of the "Population".

2) Statistic - Descriptive measure of the "Sample".

The next thing which is important for us is how we can select some data from the "Popultion" to get the "Sample" over which we want to perform our analysis. This process is called "Sampling".

Now let's see the different sampling methods:

• Probability sampling or Random Sampling - In this category of sampling, each and every individual in the "Population" is equally probable to get selected in the "Sample". There is no bias on any individual in the "Population".

• Non-probability sampling - In this category of sampling, all the individuals in the "Population" are not equally probable to get selected. There is some bias on each and every individual in the "Population".

Here we will be discussing mainly about "Random Sampling". There are various algorithms for "Random Sampling" :

• Simple Random Sampling

• Stratified sampling

• Cluster Sampling

• Systematic sampling

• Multi stage Sampling


## Lecture - 3 (Simple Random Sampling)

Let's discuss abot the first algorithm of "Random Sampling" that is "Simple Random Sampling". in this algorithm, we select our "Sample" from the "Population" in complete random manner without any kind of bias. Each and every element of the "Population" has equal chances of being selected in the "Sample".

Whenever we use any sampling algorithm, we need to make sure that the "Sample" should represent the "Population" really well. Let's try to implement this algorithm on the dataset "Startup Funding".

```python
# Maximum, Minimum, Average amount invested in any Startup (refer to column "AmountInUSD")
import pandas as pd

# Loading the dataset "Startup Funding.csv"
data = pd.read_csv('Startup Funding.csv')
data
```

Let's first try to extract some sample from the data and apply the calculations over that sample

```python
# Extracting the column containg the amount invested in USD for each Startup
amount = data['AmountInUSD']
amount
```

There are 2 problems. First of all we have "NaN" values. Secondly, these values are not of type integer (numeric). They must be string as they are having ",". So, to find the Maximum, Minimum, Average amount invested, we have to deal with these problems.

```python
# removing the "NaN" values from "amount"
amount.dropna(inplace = True)

# removing "," from "amount"
amount = amount.str.replace(',', '')

# converting the values of "amount" (string) to integer
amount = pd.to_numeric(amount)

# printing "amount"
amount
```

In the method "pandas.to_numeric()", there is one more argument "copy" which is "True" by default. This means that the method will return a copy and it will not make changes in the original dataframe. 

```python
# Maximum, Minimum, Average of the "Population"
pop_max = amount.max()
pop_min = amount.min()
pop_avg = amount.mean()

# printing Maximum, Minimum, Average of the "Population"
print(" Population Maximum = ", pop_max, "\n", "Population Minimum = ", pop_min, "\n", "Population Average = ", pop_avg)
```

To randomly select a record from a pandas dataframe, we have the method "sample()". It takes arguments : "n" (size of the sample), "random_state" (to ensure that everytime we execute the statement, we get the same random samples). We can pass any integer value to the argument "random_state".

This function by default selects "Sample" from "Population" without replacement. This means that if a record is selcted for the "Sample", it will not be selected again.

```python
# implementing sampling algorithm : Simple Random Sampling. First step is to define the sample size.
sample_size = 100

# now we have to randomly select 100 records from the "Population"
population = amount
sample = population.sample(sample_size, random_state = 1)

# Maximum, Minimum, Average of the "Sample"
sample_max = sample.max()
sample_min = sample.min()
sample_avg = sample.mean()

# printing Maximum, Minimum, Average of the "Sample"
print(" Sample Maximum = ", sample_max, "\n", "Sample Minimum = ", sample_min, "\n", "Sample Average = ", sample_avg)
```

Here we see that the set of values for "Population" and "Sample" is not exactly the same. We realize that they cannot be exactly the same. The set of values for "Sample" should ideally represent the "Population". The difference between the set of values between the "Sample" and the "Population" is called "Sampling Error".

##### Sampling Error = Parameter - Statistic

So, our goal should be to minimise the "Sampling Error" so that the "Sample" truly reflects the "Population". There are many ways with which we can reduce the "Sampling Error".

## Lecture - 4 (Stratified Sampling)

Let's talk about another sampling algorithm that is "Stratified Sampling". The word "Stratified" has been derived from the word "Strata" which means a group of elements/individuals which are having some common property. 

Here we basically group elements of the same category in one group. Suppose in a school of 1 thousand students, one way to group the students is by the "Standared" in which they are currently studying.

So, the concept behind this is, from the "Population", we have to group the elements of the "Population" in different "Strata" based on some common factor (age, gender, etc). While sampling we will pick elements from each and every "Strata" (group) using "Simple Random Sampling". 

This ensures that our sample has elements from each and every kind of possible category which is important in order to solve our problem statement. 

Suppose I want to find the "Music Taste"  of students in a school. Suppose the school has 10 thousand students. For sampling if I choose "Simple Random Sampling", then waht can happen? I suppose the sample size = 500. That means I have to select 500 students randomly from the 10 thousand students. It can happen that out of 500 students 70% students belong to "standard" 10 to 12 and 30% students belong to "standard" 7 to 9. 

"Music Taste" can differ between different age groups. The above sample will not be very good to study the music taste of the sudents. 

Hence, to solve this problem, we use "Stratified Sampling" to ensure that my sample is having data about each and every possible age group. For this we can categorise the data in different "Strata" (based on age group).

We want to find out the maximum amount of funding given to any startup in following regions -

Bangalore 

NCR (which includes New Delhi, Gurgaon and Noida)

Mumbai 

Pune 

Hydreabad 

So create your population data first by extracting those startups which are in given cities.

Then rather than using SRS for extracting sample, create 5 different stratas based on 5 regions and then apply stratified sampling on these 5 stratas to extract a sample of size 100 startups (taking 20 values from each strata)
Find and print Sampling error.

Note:

Take city name "Delhi" as "New Delhi".
Check the case-sensitiveness of cities also. That means - at some place, instead of "Bangalore", "bangalore" is given. Take city name as "Bangalore"

```python
# Getting the complete dataset
import pandas as pd
data = pd.read_csv('Startup Funding.csv')
df = data.copy()
```

```python
# Getting the "Population"
population = df.loc[(df['CityLocation'] == 'Bangalore') | (df['CityLocation'] == 'bangalore') | (df['CityLocation'] == 'Delhi') | (df['CityLocation'] == 'New Delhi') | (df['CityLocation'] == 'Gurgaon') | (df['CityLocation'] == 'Noida') | (df['CityLocation'] == 'Mumbai') | (df['CityLocation'] == 'Pune') | (df['CityLocation'] == 'Hyderabad')]
population
```

```python
# Getting the data for "Bangalore"
bangalore = df.loc[(df['CityLocation'] == 'Bangalore') | (df['CityLocation'] == 'bangalore')] 

# Getting the data for "NCR (which includes New Delhi, Gurgaon and Noida)"
ncr = df.loc[(df['CityLocation'] == 'Delhi') | (df['CityLocation'] == 'New Delhi') | (df['CityLocation'] == 'Gurgaon') | (df['CityLocation'] == 'Noida')] 

# Getting the data for "Mumbai"
mumbai = df.loc[(df['CityLocation'] == 'Mumbai')]

# Getting the data for "Pune"
pune = df.loc[(df['CityLocation'] == 'Pune')]

# Getting the data for "Hyderabad"
hyderabad = df.loc[(df['CityLocation'] == 'Hyderabad')]
```

```python
# Finding the maximum amount invested for the Population
amount = population['AmountInUSD']

# droping null values from "amount", removing "," and converting it's each element to numeric type
amount.dropna()
amount = amount.str.replace(',', '')
amount = pd.to_numeric(amount)

# maximum investment from "Population"
max_pop = amount.max()
```

```python
# performing "Stratified Sampling" and then finding the maximum amount invested
samples = []

# "strata : Bangalore"
amount = bangalore['AmountInUSD'].dropna().str.replace(',', '')
amount = pd.to_numeric(amount)
sample = amount.sample(20)
for i in sample:
    samples.append(i)
    
# "strata : NCR"
amount = ncr['AmountInUSD'].dropna().str.replace(',', '')
amount = pd.to_numeric(amount)
sample = amount.sample(20)
for i in sample:
    samples.append(i)
    
# "strata : Mumbai"
amount = mumbai['AmountInUSD'].dropna().str.replace(',', '')
amount = pd.to_numeric(amount)
sample = amount.sample(20)
for i in sample:
    samples.append(i)
    
# "strata : Pune"
amount = pune['AmountInUSD'].dropna().str.replace(',', '')
amount = pd.to_numeric(amount)
sample = amount.sample(20)
for i in sample:
    samples.append(i)
    
# "strata : Hyderabad"
amount = hyderabad['AmountInUSD'].dropna().str.replace(',', '')
amount = pd.to_numeric(amount)
sample = amount.sample(20)
for i in sample:
    samples.append(i)
    
max_sam = max(samples)
```

```python
sampling_error = max_pop - max_sam
sampling_error
```

## Lecture - 5 (Cluster Sampling)

Let's talk about another sampling algorithm that is "Cluster Sampling". Just like "Stratified Sampling", in this algorithm also we have to create different groups from the "Population". These groups are generally formed based on some common factor/property.

Let's say from our population, total "N" groups are created. Out of the "N" groups we randomly select some "n" groups as a whole. 

A major disadvantage that we can see here is that if we use "Cluster Sampling", we won't be able to get data from another possible clusters, which we have created from the "Population".

There is an advantage due to which at times we can take help of "Cluster Sampling". The advantage is the easiness to implement this algorithm (ease of collecting data). It eases our efforts of collecting data which makes it helpful in some situations where hving data from each and every possible cluster is not required.

In "Cluster Sampling", there can be multiple stages. The possible variations of the algorithm is :

• Single Stage Cluster Sampling - Here we catagorize the "Population" in different clusters and randomly select few clusters completely.

• Two Stage Cluster Sampling - Here we catagorize the "Population" in different clusters and randomly select few clusters. We apply "Simple Random Sampling" on the selected clusters to get the data.

## Lecture - 6 (Systematic Sampling)

Let's talk about another sampling algorithm that is "Systematic Sampling". It does not use random manner to select elements from "Population" to "Sample". Rather it must be using some systematic manner. 

In this algorithm, we create a sequence of our data which belongs to the "Population" and then we decide a periodic interval. According to this algorithm, a random sample, with a fixed periodic interval, is selected from the "Population". 

The major advantage of using this algorithm is that with the help of this algorithm, we can choose data from start to end in a completely systematic manner, in fixed intervals. In this situation we are making sure that we are having data about each and every possible interval (in fixed periodic interval we have defined). "Systematic Sampling" does not affect the seasonality effect on the data.
