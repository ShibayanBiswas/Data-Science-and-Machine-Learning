## Lecture - 1 (Measures in Descriptive Statistics)

Let's learn about descriptive statistics. Descriptive statistics means representing and summerizing the data quantitatively (with the help of numbers, plots, tables, etc). 

In order to represent/summerize the data with the help of some numeric value, we should be having some measures with which we can find out the number and then represent the data. We have 4 different kinds of measures with the help of which we can represent/summerize the data:

1) Measure of Central Tendency - With the help of this measure, we try to find out the central point of distribution of values. 

2) Measure of Spread/Dispersion - With the help of this measure, we try to find out how the data is spread with respect to the central value. 

3) Measure of Position - With the help of this measure, we try to find out the position where a specific data point will fall.

4) Measure of Association - With the help of this measure, we try to find out the relationship between 2 columns/variables/attributes.

## Lecture -2 (Measures of Central Tendency)

Let's talk about Measures of central Tendency. As we have discussed in the last session, it means that we are trying to find a single numeric value with which we can represent the central point of the dataset (central value about which the complete data is distributed). It has 3 measures :



#### Mean :

It means average (arithmetic mean). It is the most popular and commonly used measure. It takes into account each and every value in dataset. Sum of the deviations of each value from the mean is always zero. It can only be used with numeric data (both discrete and continuous). It is sensitive to outliers. The mean of the dataset would always change if we change any value of the data set. Since we are summing up all the values together to get it, every value of the data set contributes to its value. Median and mode may or may not change with altering a single value in the dataset.

Mean of population is denoted by "μ".

Mean of sample is denoted by "x̄".

#### Median : 

It represents exact middle point of a sorted data set. It can be used when data is having outliers. It is also used with numerical data only. To find the median, the data should be arranged first in order of least to greatest or greatest to the least value. A median is a number that is separated by the higher half of a data sample, a population or a probability distribution from the lower half. The median is different for different types of distribution. 

#### Mode :

Mode is the most frequently occurring value or category in the data set. Normally, the mode is used for categorical data where we wish to know which 
is the most common category. A data can have one or more than one mode. Mode can be used both numerical and categorical data. 

#### When to	use	which measure ?

• Mean - Data is numerical, symmetric and have no outliers.

• Median - When data is numerical and skewed or having outliers.

• Mode - Used with categorical data, can be used with numerical data also.


In case of Left Skewed/Right Skewed data we should take help of the "Median" rather than "Mean" to accurately represent the central point of the data. 

## Lecture - 3 (Measures of Spread)

The limitation of "Measure of Central Tendency" is that it does not tells us how the data is distributed (for example : deviations of individual data points from their mean). So, in this session let's discuss about "Measure of Spread". It represents the amount of dispersion in a dataset i.e. how spread out are the values. How far away the data points tend to fall from the center. This type of measure only applies to ordinal and numeric data that can be ranked. It has 5 measures :

• Range

• IQR (Interquartile range)

• Mean Absolute Deviation

• Variance

• Standard deviation

A measure of spread gives us an idea of how well the mean represents the data. When a distribution has lower variability, the values in a dataset are more 
consistent (closer to the mean and we can take the help of mean to represent the data). When the variability is higher, the data points are more dissimilar and extreme 
values become more likely (far from the mean and we cannot take the help of mean to represent the data).


#### Range :

It is the simplest measure of variation. It is the difference between the largest and smallest value.

A smaller value of range means that the data is quite consistent. This means that the majority of the data points lies very close to the mean value. A higher value of range means that the data is quite inconsistent. This means that it has extreme data points and majority of the data points are not close to the mean value.

It has some disadvantages. It does not represent actual 
data distribution. It is very sensitive to outliers. It does not consider each value of the dataset (reason for the 2 previous problems).

####  IQR (Interquartile Range) :

To understand IQR very well, we need the know the 2 terms properly : "Percentile" and "Quartile". These 2 terms are "Measures of Position". 

• **Percentile** : A percentile is defined as the percentage of values found under the specific values. Percentiles are mostly used in the ranking system. It is based on dividing up the normal distribution of the values.

• **Quartiles** : Quartiles tell us about the spread of a data set by breaking the data set into quarters, just like the median breaks it in half. Quartiles are a useful measure of spread because they are much less affected 
by outliers or a skewed dataset than the equivalent measures. Quartiles are often reported along with the median as the best choice of measure of spread and central tendency, when dealing with skewed and/or data 
with outliers.

IQR is a measure of where the majority of the values lies. IQR = 3rd Quartile - 1st Quartile.


#### Mean Absolute Deviation :

It represents the amount of variation that occurs around the mean value in dataset. It is calculated by taking average of sum of absolute difference between each value of data set and mean.

#### Variance :

The variance is a numerical value used to indicate how widely individuals in a group vary. It measures dispersion around the mean. If the values in data are spread out, the variance will be a large number. Conversely, if the values are spread closely around the mean, the variance will be a smaller number.



Problems in using variance - 

• Gives more weight to extreme scores.

• Variance is not in the same units as the values in our data set, variance is measured in the units squared.

#### Standard Deviation :

It is the measurement of average distance between each quantity and mean. 
That is, how data is spread out from mean. It is square root of variance.
A low standard deviation value indicates that the data points tend to be close to the mean, while a high value indicates that the data points are spread out over a wider range of values.


## Lecture - 4 (Measure of Position)

In this session, we will discuss the different "Measures of Position" with the help of which, we can represent the position of a data point with respect to all other data points that are available in the dataset. It has 3 measures :

• Percentile

• Quartile

• Standard Score (Z-score)


#### Percentile : 

A percentile is defined as the percentage of values found under the specific values. Percentiles are mostly used in the ranking system. It is based on dividing up the normal distribution of the values.

#### Quartiles : 

Quartiles tell us about the spread of a data set by breaking the data set into quarters, just like the median breaks it in half. Quartiles are a useful measure of spread because they are much less affected 
by outliers or a skewed dataset than the equivalent measures. Quartiles are often reported along with the median as the best choice of measure of spread and central tendency, when dealing with skewed and/or data 
with outliers. 

#### Standard Score (Z-score) :

A standard score indicates how many standard deviations an element is away from the mean. Z-score is a measure how much the data differs from mean. We can determine whether it is different enough to be significant. To calculate a Z-score, we take the individual values from the datase, subtract the mean and then divide this difference by the standard deviation. If Z-score is negative then the data point is below average. If Z-score is 0 then the data point is equal to the average. If Z-score is positive then the data point is above the average.




If the value of Z-score is less then the difference between the data point and mean is less. This implies that the data point lies in the majority section, if the distribution is normal distribution. If the value of Z-score is high then the difference between the data point and mean is high. This implies that the data point lies in outliers, if the distribution is normal distribution. We also take the help of Z-score to comapre 2 different kind of values.
