## Lecture - 3

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df)
plt.show()
```

In the above plot, we observe that the tip increases as the total bill increases. Scatter Plot is more favourable for this visualisation.

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, kind = 'line')
plt.show()
```

We can decide the type of plot we want with the help of the 'kind' parameter. It takes 2 values :- 'scatter' (default), 'line'.

We have a parameter hue (grouping variable) that produces elements of different colours. 

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, hue = 'time')
plt.show()
```

We have a parameter style (grouping variable) that produces elements of different markers. 

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, style = 'time')
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, style = 'time', hue = 'smoker')
plt.show()
```

The above graph looks quite cluttered. We can have a better means of visualisation.

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, style = 'time', col = 'smoker')
plt.show()
```

In this case when we use 'col' parameter, we get 2 plots beacase we are passing the field 'smoker' in it which has 2 values.


```python
sns.relplot(x = 'total_bill', y = 'tip', data = df, hue = 'time', col = 'smoker')
plt.show()
```

We can observe that in the second plot, the 'total_bill' and the 'tip' are more linearly related.

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = df, hue = 'time')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set()
```

```python
df = sns.load_dataset('flights')
df.head()
```

```python
sns.relplot(x = 'year', y = 'passengers', data = df)
plt.show()
```

```python
sns.relplot(x = 'year', y = 'passengers', data = df, kind = 'line')
plt.show()
```

From the above plot, we observe that the number of passengers have increased gradually every year.

```python
sns.relplot(x = 'year', y = 'passengers', data = df, kind = 'line', hue = 'month')
plt.show()
```

In the above plot we observe 12 different lines for 12 months.

The 'style' parameter in this case is available for only 6 types of values, present in a field. If a field containing more than 6 type of values is passed, we get an error. For example - If field 'month' is passed in 'style', we get an error.

```python
sns.relplot(x = 'year', y = 'passengers', data = df, kind = 'line', col = 'month')
plt.show()
```

We obtain 12 different plots for 12 different months in the above case where each plot shows trend between 'passengers' and 'year'.

```python
sns.lineplot(x = 'year', y = 'passengers', data = df)
plt.show()
```

## Lecture - 5

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df)
plt.show()
```

The 'jitter' parameter helps us to control the randomness of the points in the plot.

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df, jitter = 1)  # Default value of jitter is 1
plt.show()
```

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df, jitter = 0.25)
plt.show()
```

In the above plot we observe that the area in which the points are distributed has increased.

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df, jitter = 0.75)
plt.show()
```

In the above plot the points have been distributed randomly.

Passing a lower value to te 'jitter' parameter means I do not want much randomness or deviation of the points in the plot.

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df, jitter = 0.005)
plt.show()
```

Swarm plot is similar to strip plot. The only difference is that it does not allow overlapping of markers.

Although it works well for relatively small datasets because sometimes it takes a lot of computation to arrange large data sets such that they are not overlapping.

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df, kind = "swarm")
plt.show()
```

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df, kind = "swarm", hue = "smoker")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df, kind = "box")
plt.show()
```

From the above plot we observe that when the customer is male, the bill is genarally above average. When the customer is female then the above average bill is lesser compared to that of a male customer.

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df, kind = "box", hue = "smoker")
plt.show()
```

From the above plot we observe that the average bill for male customers (smokers) is more and in case of female customers the average bill is relatively same.

In case of male customers (non smokers) there are more number of outliers.

```python
sns.catplot(x = 'sex', y = 'total_bill', data = df, kind = "boxen", hue = "smoker")
plt.show()
```

## Lecture - 7

If we want to get a sense about the density distribution of the dataset then we should use violin plot.

KDE (Kernel Density Estimation) tries to estimate (based on the given dataset) that probalistically how many data points will lie at a particular value on the Y - axis. That means hoe dense the data points will be at a particular value on the Y - axis. 

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set()
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = df, kind = 'violin')
plt.show()
```

From the above plot we can infer the density distribution of the points for a value on the Y - axis.

```python
sns.catplot(x = 'species', y = 'sepal_width', data = df, kind = 'violin')
plt.show()
```
