## Lecture - 1 (What is Hypothesis Testing)

In this lecture, we will be talking about Hypothesis Testing. First let's discuss what exactly a Hypothesis is? what exactly a Hypothesis Testing is?

Let's imagine you are working as a software developer. You have suggested a new feature that will ultimately improve the customer experience to a next level. So, the idea sounds quite interesting to everyone. Implementation of the idea will require a lot of resources from the company (money, time, etc). So, before proceeding with the idea, the comapny wants to be 100% sure that the feature is worth adding. So, the idea here is the **Hypothesis**.


Hypotheses are always statements or assumptions about the population which we want to verify. For example :

&nbsp; &nbsp; &nbsp; • Will I improve my grades if I spend 4 hrs studying daily ?

&nbsp; &nbsp; &nbsp; • If breakfast helps children perform better in schools ?

&nbsp; &nbsp; &nbsp; • If the average time that college students spend studying each week is 20 hours per week ?


A Hypothesis is an idea made from limited evidence, and is a starting point of further investigation. This is where you can use the sample data to answer the research questions.

A Famous saying - “A fact is a simple statement that everyone believes. It is innocent, unless found guilty. A hypothesis is a novel suggestion that no one wants to believe. It is guilty,
until found effective.”

**Hypothesis Testing** is an inferential procedure that uses sample data to evaluate the credibility of a hypothesis about a population. It is a kind of statistical inference that involves asking a question, collecting data, and then examining what the data tells us about how to proceed.

A hypothesis test is a rule that specifies whether to accept or reject a claim about a
population depending on the evidence provided by a sample of data.


## Lecture - 2 (Null & Alternate Hypothesis)

Whenever we are drawing any kind of inferences in **Hypothesis Testing**, we need to have 2 hypotheses. These 2 hypotheses are basically termed as :

&nbsp; &nbsp; &nbsp; • The null hypothesis (H<sub>0</sub>)

&nbsp; &nbsp; &nbsp; • The alternative hypothesis (H<sub>1</sub> or H<sub>A</sub>)

#### Null Hypothesis (H<sub>0</sub>) :

&nbsp; &nbsp; &nbsp; • In hypothesis testing, we begin by making a tentative assumption about population parameter. This tentative assumption is called null hypothesis.

&nbsp; &nbsp; &nbsp; • A statement about the population parameter i.e. the statement which we want to test.

&nbsp; &nbsp; &nbsp; • Can include : =, >= , <=

#### Alternate Hypothesis (H<sub>1</sub> or H<sub>A</sub>) :

&nbsp; &nbsp; &nbsp; • A statement that directly contradicts Null Hypothesis.

&nbsp; &nbsp; &nbsp; • Can include : ≠, >, <

#### Examples :

• A school claims that on an average their students get at least 70% marks :

&nbsp; &nbsp; &nbsp; • **Claim** - Average marks >= 70%
    
&nbsp; &nbsp; &nbsp; • **Counterclaim** - Average marks < 70%
    
• According to a survey, average number of hours spent by phd students in their research work is more than 10 hours per day :

&nbsp; &nbsp; &nbsp; • **Claim** - Average hours > 10 hrs
    
&nbsp; &nbsp; &nbsp; • **Counterclaim** - Average hours <= 10 hrs
    
• A company states that average life of their car tyres is 36 months :

&nbsp; &nbsp; &nbsp; • **Claim** - Average life = 36 months
    
&nbsp; &nbsp; &nbsp; • **Counterclaim** - Average life ≠ 36 months

## Lecture - 3 (Significance Level)

Now after defining the Null and Alternate Hypothesis, next we need to define is the decision criteria based on which we are going to accept or reject the Null Hypothesis. 

In an ideal situation, we should reject the Null Hypothesis if it is false and accept the Null Hypothesis, if it is true. Here the decision criteria is known as **Significance Level**.

Significance Level is denoted by Alpha (α). It refers to the degree of significance in which we accept or reject the Null Hypothesis. It is the probability of rejecting the null hypothesis, if it is true. Typical values for Alpha (α) are - 0.01, 0.05, 0.1. The values of Alpha (α) basically tells us about the chances (in %) that the result is incorrect.

The choice of Alpha (α) is determined by the context you are operating in but 0.05 or 5% is the most commonly used value.

Alpha (α) = 0.05 means, your output should be 95% confident to give similar kind of result in each sample.

Based on the level of significance, we make a decision to accept the Null or Alternate hypothesis.

Suppose there is a tyre manufacturing comapany which manufactures tyres for different cars. The company claims that the lifetime of their tyres is 36 months on an average. Now we need to verify the claim. In order to verify the claim, we need Hypothesis Testing. 

To check if the claim is correct or not, we can find out the average lifetime of the tyres in the "Sample". We cannot achieve 100% accuracy. If the company claims that the lifetime of their tyres is 36 months on an average then it does not imply that each and evry tyre manufactured will have a lifetime of 36 months. There will be some error. 

If the average lifetime of the tyres in the "Sample" is very close to the comapny's claim then we can say that the company's claim is correct. If the average lifetime of the tyres in the "Sample" is not close to the comapny's claim then we can say that the company's claim is incorrect.

Hence, we can decide whether to accept/reject the Null Hypothesis on the basis of the average lifetime of the tyres in the "Sample". We don't know for which range of values (average lifetime of tyres) we should consider the comapany's claim or not. In order to clear this confusion we define **Critical Value**.

In statistical hypothesis testing, the critical values of a statistical test are the boundaries of the acceptance region of the test. If a test statistic on one side of the critical value results in accepting the Null Hypothesis, a test statistic on the other side will result in rejecting the Null Hypothesis.

#### Steps for using critical value in hypothesis testing :

&nbsp; &nbsp; &nbsp; • Calculate the test statistic

&nbsp; &nbsp; &nbsp; • Calculate critical values based on significance level (alpha)

&nbsp; &nbsp; &nbsp; • Compare test statistic with critical values


#### Rejection Region : 

• Rejection region represents a set of values for the test statistic, for which the Null Hypothesis is rejected.

• If the observed test statistic is in the critical region then we reject the Null Hypothesis and accept the Alternative Hypothesis.

#### Type of Tests :

According to Central Limit Theorem, if we plot the Sampling Distribution for every "Population", we get a Normal Distribution curve. In order to define the acceptance/rejection of the Null Hypothesis, we have 3 different type of tests that are available with us : 

&nbsp; &nbsp; &nbsp; • **Left tailed test** : The critical region is in the extreme left region (tail) under the curve.

&nbsp; &nbsp; &nbsp; • **Right tailed test** : The critical region is in the extreme right region (tail) under the curve.

&nbsp; &nbsp; &nbsp; • **Two tailed test** : The critical region of a distribution is two-sided and tests whether a sample is greater than or less than a certain range of values.

Depending on the nature of alternate hypothesis, nature of statistical test will be used. 

#### Rules :

&nbsp; &nbsp; &nbsp; • If there is less than sign in alternate hypothesis, then we use left tail test.

&nbsp; &nbsp; &nbsp; • If there is greater than sign in alternate hypothesis then we use right tail test.

&nbsp; &nbsp; &nbsp; • If there is not equal sign in alternate hypothesis then we use two tail test.


## Lecture - 5 (Errors in Hypothesis Testing)

Till now, we have understood that how can we perform Hypothesis Testing to get some inference about the "Population" based on data of the "Sample". We know that any kind of test cannot be 100% accurate. There are always some chances of error. In Hypothesis Testing also, we might incur some kind of error. Let's try to understand what these errors are.

We have 2 claims to be tested - Null Hypothesis and Alternate Hypothesis. Only one of them can be true. Ideally we should not reject the null hypothesis when it is true and we should 
reject it when it is false (or alternate hypothesis is true).


There are 2 types of errors - 

&nbsp; &nbsp; &nbsp; • **Type 1** error 

&nbsp; &nbsp; &nbsp; • **Type 2** error

#### Type 1 Error :

It occurs when you reject a true null hypothesis. It is also called a false positive. Bigger the rejection region, higher are the chances of making Type 1 error. The probability of making this error is alpha, the level of significance. An α = 0.05 indicates that you are willing to accept a 5% chance that you are wrong when you reject the null hypothesis. 

#### Type 2 Error :

It occurs when you accept a false null hypothesis. It is also called a false negative. The probability of making this error is denoted by Beta (β). We should also mention that the probability of rejecting a false null hypothesis is (1 - β). This is the researchers goal to reject a false null hypothesis. Therefore (1 - β) is called **The Power of the test**.

As we increase the rejection region, the probablity of making the Type 1 error increases but the probablity of making the Type 2 error decreases. As we decrease the rejection region, the probablity of making the Type 1 error decreases but the probablity of making the Type 2 error increases.

Hence, while we are adjusting the Critical Value, the chances of one error decreases and the chances of the other error increases. Hence, we need to take a trade-off where the chances of getting Type 1 error and Type 2 error is minimized.
